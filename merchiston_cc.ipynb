```python
f = folium.Figure(width=800, height=1200)
tiles = "openstreetmap"
m = folium.Map(location=merchiston_tower, tiles=tiles, zoom_start=14)
```

```python
# cc = path("da44854d46c44a76b9b1160fc609738a_25_community_councils.geojson.gz")
cec_cc = "http://data.edinburghcouncilmaps.info/datasets/da44854d46c44a76b9b1160fc609738a_25.geojson"
cc_data = gpd.read_file(cec_cc)

# with gzip.open(cc, "rb") as f:
#     cc_data = gpd.read_file(f)

# ccfilter = (cc_data["LABEL"] == "Gilmerton/Inch") | (
#     cc_data["LABEL"] == "Liberton and District"
# )

ccfilter = (cc_data["LABEL"] == "Merchiston")

mcc = cc_data[ccfilter]
mcc
```

```python
mcc.to_file("mcc_boundaries.json", driver="GeoJSON")
```

```python
cc_geo = folium.features.GeoJson(
    mcc,
    style_function=lambda x: {"color": "red", "fillOpacity": 0.1, "weight": 2},
    tooltip=folium.features.GeoJsonTooltip(
        fields=["LABEL"], aliases=["Community Council"], labels=True, sticky=False
    ),
)
m.add_child(cc_geo)
```
